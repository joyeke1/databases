```python
# Created an inner join to produce the first name of customers and the amount of their payments only if their payments were equal to 1.99
```

```python
!psql -c "select first_name, amount from pagila.payment join pagila.customer on payment.customer_id = customer.customer_id where payment.amount = 1.99 order by payment.amount limit 9"
```

```python
#Used an inner join to produce the film titles of film's in english
```

```python
!psql -c "select title, name from pagila.film join pagila.language on film.language_id= language.language_id where language.name = 'English' order by film.title limit 10"
```

```python
# used an inner join to bring forward the film_id whose rental_id was between 1 and 10
```

```python
!psql -c "select film_id, rental_id from pagila.inventory join pagila.rental on inventory.inventory_id = rental.inventory_id where rental_id between 1 and 10 order by rental_id "
```

```python
# Created an outer join to show the city's in the country of Peru
```

```python
!psql -c "select c.country, a.city from pagila.country c full join pagila.city a on c.country_id = a.country_id where c.country = 'Peru'order by a.city"
```

```python
#Used an outer join to show the first 9 tittle films which are in english or french
```

```python
!psql -c "select f.title, l.name from pagila.film f full join pagila.language l on l.language_id= f.language_id where l.name = 'English' or l.name = 'French 'order by f.title limit 9"
```

```python
# Used an outer join to show the actor_id and category_id in films which have an actor_id between 1 and 2
```

```python
!psql -c "select f.actor_id, c.category_id from pagila.film_actor f full join pagila.film_category c on f.film_id = c.film_id where f.actor_id between 1 and 2 order by f.actor_id limit 10"
```

```python
#Created a self join to show the payment_id and customer_id where the same amount of payment was made
```

```python
!psql -c "select a.payment_id, b.customer_id from pagila.payment a join pagila.payment b on a.amount = b.amount limit 9"
```
